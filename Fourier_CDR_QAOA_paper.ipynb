# QAOA mitigation experiments

Code to generate figures and results relating to spectral error mitigation of QAOA experiments.

The code is split into two parts: 1. raw data generation (requires access to quantum computer) and 2. mitigation (from loaded raw data file)

Own simplified implementation of Clifford Data Regression. Original paper at https://arxiv.org/pdf/2005.10189.pdf

# 1. Raw data generation

### Circuit construction

```python
def qaoa(params_b, params_c, n_qubits, layers, edges):
    assert len(params_b) == len(edges) * layers
    assert len(params_c) == n_qubits * layers
    
    params_b_iter = iter(params_b)
    params_c_iter = iter(params_c)
    
    ansatz = Circuit(n_qubits)
    for _ in range(n_qubits):
        ansatz.H(_)
        
    for l in range(layers):
        for _ in range(len(edges)):
            ansatz.ZZPhase(next(params_b_iter), *edges[_])
        for _ in range(n_qubits):
            ansatz.Rx(next(params_c_iter), _)
            
    ansatz.measure_all()
    return ansatz
```

```python
# Some randomly renerated 3-regular graphs

n_qubits = 8

if n_qubits == 8:
    edges = [(0, 1), (0, 7), (0, 2), (1, 4), (1, 5), (6, 7), (6, 4), (6, 3), (7, 5), (4, 2), (5, 3), (2, 3)]
    
elif n_qubits == 16:
    edges = [(8, 13), (8, 10), (8, 3), (13, 12), (13, 0), (5, 6), (5, 2), (5, 1), (6, 14), (6, 11), (0, 14),
             (0, 3), (14, 9), (7, 12), (7, 11), (7, 3), (12, 15), (1, 15), (1, 4), (15, 11), (2, 9), (2, 10),
             (4, 9), (4, 10)]

G = nx.from_edgelist(edges)
nx.draw(G)
```

```python
layers = 1

syms_b = symbols(f"b0:{len(edges) * layers}")
syms_c = symbols(f"c0:{n_qubits * layers}")

circ = qaoa(syms_b, syms_c, n_qubits, layers, edges)
render_circuit_jupyter(circ)
```

### Data generation

```python
backend = "ibmq_guadalupe"
noisy_backend = IBMQBackend(backend)
```

```python
noiseless_backend = AerBackend(simulation_method="statevector")
```

```python
def generate_circuit_from_ansatz(ansatz: Circuit, backend: IBMQBackend, var_params_range, fix_params):
    circs = []
    
    for i in var_params_range:
        for j in var_params_range:
            c = ansatz.copy()
            
            # Problem unitary
            b_params = np.concatenate([np.repeat(fix_params[1], len(edges)), np.repeat(j, len(edges))])
            sym_map_b = dict(zip(syms_b, b_params))
            c.symbol_substitution(sym_map_b)
            
            # Mixing unitary
            c_params = np.concatenate([np.repeat(fix_params[0], n_qubits), np.repeat(i, n_qubits)])
            sym_map_c = dict(zip(syms_c, c_params))
            c.symbol_substitution(sym_map_c)
            
            backend.compile_circuit(c)
            circs.append(c)
            
    return circs

def generate_random_clifford_params(non_cliff_params, samples=1, num_non_cliffs=2):
    params = []
    cliffs = np.array([0.5, 1., 1.5])  # Clifford angles
    
    for s in range(samples):
        p = non_cliff_params.copy()
        ixs = np.random.choice(range(len(p)), size=len(p)-num_non_cliffs, replace=False)  # Indices of gates to be replaced with Clifford gates
        for ix in ixs:
            smallest_difference_ix = abs(cliffs - p[ix]).argmin()
            closest_cliff = cliffs[smallest_difference_ix]
            p[ix] = closest_cliff
        params.append(p)
    return params

def generate_clifford_circuits(ansatz: Circuit, backend: IBMQBackend, var_params_range, fix_params):
    circs = []
    
    for i in var_params_range:
        for j in var_params_range:
            b_params = np.concatenate([np.repeat(fix_params[1], len(edges)), np.repeat(j, len(edges))])
            c_params = np.concatenate([np.repeat(fix_params[0], n_qubits), np.repeat(i, n_qubits)])
            
            params = np.concatenate([b_params, c_params]) 
            params_cliff = generate_random_clifford_params(params, samples=1, num_non_cliffs=10)
            
            for p in params_cliff:
                c = ansatz.copy()
                
                # Problem unitary
                b_params_cliff = p[:layers * len(edges)]
                sym_map_b = dict(zip(syms_b, b_params_cliff))
                c.symbol_substitution(sym_map_b)
                
                # Mixing unitary
                c_params_cliff = p[layers * len(edges):]
                sym_map_c = dict(zip(syms_c, c_params_cliff))
                c.symbol_substitution(sym_map_c)
                
                backend.compile_circuit(c)
                circs.append(c)
            
    return circs
```

```python
# Parameters of experiments

N = 6  # Max frequency mode to resolve
n = 2 * N + 1  # Sampling rate

delta = 0.5 / n  # Small offset to prevent compilation to identity
x = linspace(0, 2, n, endpoint=False) + delta

fix_params = [1./4., 1./4.]  # Fixed parameters of layers
```

```python
circs = generate_circuit_from_ansatz(circ, noisy_backend, x, fix_params)
```

```python
circs_cliff_exp = generate_clifford_circuits(circ, noisy_backend, x, fix_params)
```

```python
circs_cliff_sym = generate_clifford_circuits(circ, noiseless_backend, x, fix_params)
```

```python
circs_exact = generate_circuit_from_ansatz(circ, noiseless_backend, x, fix_params)
```

```python
def get_results(backend: IBMQBackend, circs, shots=8192):
    if backend == AerBackend():
        shots = 1e5
        
    res_handle = backend.process_circuits(circs, n_shots=shots)
    return backend.get_results(res_handle)
```

# 2. Mitigation

### Load data

```python
def load_file(name):
    file = open(name, 'rb')
    res_dict = pickle.load(file)
    file.close()
    return res_dict

def measure_subset(d, indices):
    r = {}
    for k in d.keys():
        k_ary = np.array(k)
        k_ = tuple(k_ary[indices])
        try:
            r[k_] += d[k]
        except KeyError:
            r[k_] = d[k]
        
    return r

def hamiltonian(x):
    return sum(expectation_from_counts(measure_subset(x, list(e))) for e in edges)

def get_hamiltonian(res_dict):
    return np.array([hamiltonian(r) for r in res_dict])
```

```python
# Parameters of experiment

M = 2  # Number of independent parameters (i.e. dimensionality of landscape)
N = 6  # Max frequency mode to resolve
n = 2 * N + 1  # Sampling rate

delta = 0.5 / n  # Small offset to prevent compilation to identity
x = linspace(0, 2, n, endpoint=False) + delta

xf = fftfreq(n, 1./n)

results_folder = "_results"

# Select experiment file to load

# Experiment A
name = "ibmq_guadalupe_20210823-194015_QAOA_CDR_simpler_8q_1p"
name_exact = "aer_backend_20210823-194015_QAOA_CDR_simpler_8q_1p"

# Experiment B
# name = "ibmq_guadalupe_20210823-185100_QAOA_CDR_simpler_8q_2p"
# name_exact = "aer_backend_20210823-185100_QAOA_CDR_simpler_8q_2p"

# Experiment C
# name = "ibmq_sydney_20210823-162832_QAOA_CDR_simpler_8q_1p"
# name_exact = "aer_backend_20210823-194015_QAOA_CDR_simpler_8q_1p"

# Experiment D
# name = "ibmq_sydney_20210823-171110_QAOA_CDR_simpler_8q_2p"
# name_exact = "aer_backend_20210823-185100_QAOA_CDR_simpler_8q_2p"

# Experiment E
# name = "ibmq_guadalupe_20210824-140520_QAOA_CDR_simpler_16q_1p"
# name_exact = "aer_backend_20210824-140520_QAOA_CDR_simpler_16q_1p"


res_dict = load_file(f"{results_folder}/{name}.pkl")
res_dict_cliff_exp = load_file(f"{results_folder}/{name}_cliff_exp.pkl")
res_dict_cliff_sym = load_file(f"{results_folder}/{name_exact}_cliff_sym.pkl")
res_dict_exact = load_file(f"{results_folder}/{name_exact}.pkl")
```

```python
obj = get_hamiltonian(res_dict)
obj_cliff_exp = get_hamiltonian(res_dict_cliff_exp)
obj_cliff_sym = get_hamiltonian(res_dict_cliff_sym)
obj_lst_exact = get_hamiltonian(res_dict_exact)
```

## 2.1 Basic CDR

```python
# Linear regression model with intercept
reg = linear_model.LinearRegression(fit_intercept=True)

# Linear regression on real space data
x = obj_cliff_sym
y = obj_cliff_exp

# Exclude outliers
mask = (abs(x) > 0.05) & (abs(y) < 0.75) & (abs(y) > 0.05)
# mask = (abs(x) > 0.2) & (abs(y) < 0.75) & (abs(y) > 0.2)
y = y[mask]
x = x[mask]

# Fit
reg.fit(y.reshape(-1,1), x.reshape(-1,1))

# Get coefficients of fitted model
scaling = reg.coef_[0,0]
intercept = float(reg.intercept_)
```

```python
pred = reg.predict(y.reshape(-1,1))
plt.scatter(y, x)
plt.plot(y, pred);
```

```python
a = obj.reshape((n, n))
C_ = fftn(a)

# Revert phase shift
phase = np.pi * delta
phase_vec = np.exp(-1j * xf * phase)
phase_array = np.outer(phase_vec, phase_vec)
C = C_ * phase_array / n ** M
```

```python
# Plot landscape in Fourier space
X, Y = meshgrid(xf, xf)
Z = np.real(C)
iZ = np.imag(C)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, Z)
ax.scatter(X, Y, iZ);
```

```python
# Do the same for exact simulator run
a_exact = np.array(obj_lst_exact).reshape((n, n))
C_exact_ = fftn(a_exact)

phase_exact = np.pi * delta
phase_vec_exact = np.exp(-1j * xf * phase_exact)
phase_array_exact = np.outer(phase_vec_exact, phase_vec_exact)
C_exact = C_exact_ * phase_array_exact / n ** M
```

```python
# Simple rescaling of fourier coeffs
shift = np.zeros(C.shape)
shift[0, 0] = intercept
C_cdr = C * scaling + shift
```

## 2.2 CDR -> Filtering

```python
mask_1 = (abs(xf) <= 2).astype(int)
mask_2 = (abs(xf) <= 5).astype(int)
mask_ft = np.outer(mask_1, mask_2)
```

```python
C_filter_after_cdr = C_cdr * mask_ft
```

### 2.3 CDR -> Thresholding

```python
norms = (C * C.conj()).real.flatten()

B = 10  # Thresholding constant
cutoff = B * np.median(norms)
C_thres = C.flatten()
C_thres[norms < cutoff] = 0
C_thres = C_thres.reshape((n, n))

x = np.arange(len(norms))
plt.bar(x, np.sort(norms))
plt.plot(x, np.repeat(cutoff, len(norms)), color='r');
```

```python
C_thres_after_cdr = C_thres * scaling * mask_ft + shift
```

# Final results plotting

```python
# Complex dot
def cdot(a, b):
    return np.sum(a.conjugate() * b)

# Cosine similarity
def cos_sim(a, b):
    return cdot(a.flatten(), b.flatten()).real / (np.linalg.norm(a) * np.linalg.norm(b))

# Euclidean distance
def dist(a, b):
    return np.linalg.norm(a - b)
```

```python
# "Clean" exact simulator results from numerical errors
C_exact_clean = C_exact.copy()
C_exact_clean[np.abs(C_exact_clean) < 1e-2] = 0 
```

```python
names = ["raw",
         "CDR",
         "CDR->filtering",
         "CDR->thresholding"]

results =  [C,
            C_cdr,
            C_filter_after_cdr,
            C_thres_after_cdr]
```

```python
fig = plt.figure(figsize=(20,10))
ax0 = fig.add_subplot(211)
ax1 = fig.add_subplot(212)

y0 = [cos_sim(r, C_exact_clean) for r in results]
ax0.bar(names, y0, alpha=0.8)
ax0.set_title("1 - Cosine similarity")

y1 = [dist(r, C_exact_clean) for r in results]
ax1.bar(names, y1, alpha=0.8)
ax1.set_title("Euclidean distance");
```

### Plotting Fourier coefficients, before and after mitigation

```python
def to_color(z):
    offset = np.mod(z + np.pi, 2 * np.pi)
    fracs = offset/(2 * np.pi)
    norm = colors.Normalize(0, 1)
    return cm.jet(norm(fracs.tolist()))

def bar3d(ax, C, title=""):
    X_, Y_ = meshgrid(xf, xf)
    X, Y = X_.ravel(), Y_.ravel()
    width = depth = 0.5
    bottom = np.zeros_like(X)
    color = to_color(np.angle(C).ravel())
    color[np.abs(C).ravel() <= 1e-8] = 0.5  # Set small coeffs to gray
    ax.bar3d(X, Y, bottom, width, depth, np.abs(C).ravel(), color=color)
    ax.set_title(title, fontsize=20)
```
